### Exercise 2 : Matplotlib's Object Oriented API
---

Matplotlib have an object-oriented API used to create more complex figures. Code is a
little more complicated, but the advantage is that we now have full control of where the plot axes. Here‚Äôs a good read when exploring matplotlib

https://github.com/rougier/matplotlibtutorial

#### 1) Create the same chart as 1.1 using the new method: instantiate figure objects and then call methods or attributes from that object.

```python
x = np.arange(0,5.5,0.5)
y = x**2
fig , axes = plt.subplots()
axes.plot(x,y)
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("Curb of the function f(x) = x^2")
plt.show()
```

#### 2) Create the figure below (the blue curve represents $y=x^2$ while the red curve represents $x=y^2$)

```python
fig = plt.figure(figsize=(10,10))
main_subplot = fig.add_axes([0.1,0.1,0.8,0.8])
main_subplot.plot(y**2,x, color="red")
main_subplot.set_xlabel("x")
main_subplot.set_ylabel("y^2")
main_subplot.set_title("Curb of the function f(x) = y^2")
nested_subplot = fig.add_axes([0.48,0.25,0.4,0.3])
nested_subplot.plot(x**2,y, color="blue")
nested_subplot.set_xlabel("y")
nested_subplot.set_ylabel("x^2")
nested_subplot.set_title("Curb of the function f(x) = x^2")
fig.show()
```

#### 3) Explore colors and markers style. Create a new chart, such as x goes from 0 to 10, and $y=x+index$.

*index* is the number of curves you're going to draw. For example, the first curve will match $y=x+0$, the second $y=x+1$, etc...

Create multiple combinations of differents colors, markers & marker size in this graphic. Here's an example using three curves:

```python
x = np.arange(0,10,1)
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,1,1])
axes.plot(x,x+0, color="red", marker="o", markersize=10)
axes.plot(x,x+1, color="blue", marker="+", markersize=10)
axes.plot(x,x+2, color="green", marker="s", markersize=10)
plt.show()
```

#### 4) Create the same subplots you've created in 1.4 using matplotlib API (use `plt.subplots()`

a. Add a title for each plot ‚Äòtitle 1‚Äô, ‚Äòtitle 2‚Äô, etc.

b. Add an x axis and y axis label for each plot (for example, x-axis: X, y-axis: f(x)= x^2

c. Use figsize to make it clearer

d. Show legend on chart

e. Differentiate between each graph using a different color and markers




```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15,15))
axes[0,0].plot(x,x**2, color="red", marker="o", markersize=10)
axes[0,0].set_title("Curb of the function f(x) = x^2")
axes[0,0].set_xlabel("x")
axes[0,0].set_ylabel("x^2")
axes[0,1].plot(x,x**3, color="blue", marker="+", markersize=10)
axes[0,1].set_title("Curb of the function f(x) = x^3")
axes[0,1].set_xlabel("x")
axes[0,1].set_ylabel("x^3")
axes[1,0].plot(x,np.log(x), color="green", marker="s", markersize=10)
axes[1,0].set_title("Curb of the function f(x) = log(x)")
axes[1,0].set_xlabel("x")
axes[1,0].set_ylabel("log(x)")
axes[1,1].plot(x,np.exp(x), color="yellow", marker="*", markersize=10)
axes[1,1].set_title("Curb of the function f(x) = e^x")
axes[1,1].set_xlabel("x")
axes[1,1].set_ylabel("e^x")
fig.legend(["x^2","x^3","log(x)","e^x"])
plt.show()
```
